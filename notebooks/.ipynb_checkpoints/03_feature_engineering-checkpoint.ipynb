# Import & Load

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/diamonds_clean.csv')
print('Data loaded:', df.shape)
df.head()
```

# Convert Price USD to INR

```python
# Using fixed conversion rate
 
USD_TO_INR = 84.0

df['price_inr'] = df['price'] * USD_TO_INR

print(f'USD to INR conversion rate: {USD_TO_INR}')
print(f'Sample prices:')
print(df[['price', 'price_inr']].head())
```

#  Create New Features

```python
# Volume of the diamond
df['volume'] = df['x'] * df['y'] * df['z']

# Price per carat
df['price_per_carat'] = df['price'] / df['carat']

# Dimension ratio
df['dimension_ratio'] = (df['x'] + df['y']) / (2 * df['z'])

# Carat category
def carat_category(c):
    if c < 0.5:
        return 'Light'
    elif c <= 1.5:
        return 'Medium'
    else:
        return 'Heavy'

df['carat_category'] = df['carat'].apply(carat_category)

print('New features created ')
print(df[['carat', 'volume', 'price_per_carat', 'dimension_ratio', 'carat_category']].head())
```

# Carat Category Distribution

```python
print('Carat Category Counts:')
print(df['carat_category'].value_counts())
```

#  Apply Log Transformation on Skewed Features

```python
# Apply log1p on highly skewed columns
skewed_cols = ['carat', 'price', 'price_inr', 'y', 'z', 'volume', 'price_per_carat']

for col in skewed_cols:
    df[f'log_{col}'] = np.log1p(df[col])
    print(f'log_{col} created ')
```

# Verify Skewness Improved

```python
print('--- Skewness Before vs After Log Transform ---\n')
for col in ['carat', 'price', 'y', 'z']:
    original_skew = df[col].skew()
    log_skew = df[f'log_{col}'].skew()
    print(f'{col:<10} Before: {original_skew:>7.4f}  →  After: {log_skew:>7.4f}')
```

# Encode Categorical Columns

```python
# Ordinal encoding — these have a natural order
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']  # D is best
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']  # IF is best

df['cut_encoded'] = df['cut'].map({v: i for i, v in enumerate(cut_order)})
df['color_encoded'] = df['color'].map({v: i for i, v in enumerate(color_order)})
df['clarity_encoded'] = df['clarity'].map({v: i for i, v in enumerate(clarity_order)})

print('Encoding done ')
print(df[['cut', 'cut_encoded', 'color', 'color_encoded', 'clarity', 'clarity_encoded']].head())
```

#  Final Feature Overview

```python
print('Final DataFrame Shape:', df.shape)
print('\nAll Columns:')
for col in df.columns:
    print(f'  → {col}')
```

# Save Engineered Dataset

```python
df.to_csv('../data/diamonds_engineered.csv', index=False)
print('Engineered dataset saved ')
print('Shape:', df.shape)
```

# Correlation Matrix Feature Selection

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['log_carat', 'depth', 'table', 'x', 'y', 'z', 
                  'volume', 'price_per_carat', 'dimension_ratio',
                  'cut_encoded', 'color_encoded', 'clarity_encoded',
                  'log_price_inr']

corr = df[numerical_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Identify Highly Correlated Features

```python
print('--- Correlation with Target (log_price_inr) ---\n')

target_corr = corr['log_price_inr'].drop('log_price_inr').sort_values(ascending=False)

for feat, val in target_corr.items():
    if abs(val) >= 0.7:
        flag = '🔴 Very High'
    elif abs(val) >= 0.5:
        flag = '🟡 High'
    elif abs(val) >= 0.3:
        flag = '🟢 Moderate'
    else:
        flag = '⚪ Low'
    print(f'{feat:<25} correlation: {val:>7.4f}  {flag}')
```

#  Select Final Features

```python
# Drop low correlation and highly redundant features
# Keep features with meaningful correlation to target

selected_features = [
    'log_carat',          # Very high correlation
    'cut_encoded',        # Ordinal quality feature
    'color_encoded',      # Ordinal quality feature  
    'clarity_encoded',    # Ordinal quality feature
    'depth',              # Physical feature
    'table',              # Physical feature
    'x', 'y', 'z',        # Dimensions
    'volume',             # Derived — high correlation
    'price_per_carat',    # Derived — high correlation
    'dimension_ratio'     # Derived feature
]

print('Selected Features for Modeling:')
for i, f in enumerate(selected_features, 1):
    print(f'  {i:>2}. {f}')

print(f'\nTotal features selected: {len(selected_features)}')
```
