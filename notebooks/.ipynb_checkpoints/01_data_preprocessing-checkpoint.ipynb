# Import Libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


print('Libraries imported Successfully')
```

# Load Dataset

```python
# Load diamonds dataset directly from seaborn (no download needed)
df = sns.load_dataset('diamonds')

# Save raw copy to data folder
df.to_csv('../data/diamonds.csv', index=False)

print(f'Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns')
df.head()
```

#  Basic Inspection

```python
print('Shape:', df.shape)
print('\n--- Data Types ---')
print(df.dtypes)
```

# Statistical Summary

```python
df.describe()
```

# Check Missing Values

```python
print('--- Missing Values ---')
print(df.isnull().sum())
print(f'\nTotal missing: {df.isnull().sum().sum()}')
```

# Check Zero Values in x, y, z

```python
# Zero is physically impossible for a real diamond dimension
print('Zero values in x:', (df['x'] == 0).sum())
print('Zero values in y:', (df['y'] == 0).sum())
print('Zero values in z:', (df['z'] == 0).sum())
```

# Fix Zero Values

```python
# Replace zeros with NaN then impute with median
df[['x', 'y', 'z']] = df[['x', 'y', 'z']].replace(0, np.nan)

for col in ['x', 'y', 'z']:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)
    print(f'{col} — imputed with median: {median_val:.4f}')
```

# Check & Drop Duplicates

```python
print('Duplicate rows:', df.duplicated().sum())
df = df.drop_duplicates()
print('Shape after dropping duplicates:', df.shape)
```

# Check Categorical Value Counts

```python
for col in ['cut', 'color', 'clarity']:
    print(f'\n--- {col} ---')
    print(df[col].value_counts())
```

# Save Clean Dataset

```python
df.to_csv('../data/diamonds_clean.csv', index=False)
print('Clean dataset saved ')
print('Final shape:', df.shape)
```

```python
# Final Summary

print('='*40)
print('   PREPROCESSING SUMMARY')
print('='*40)
print(f'Total rows after cleaning : 53,794')
print(f'Zero values fixed (x,y,z) : 35')
print(f'Duplicates removed        : 146')
print(f'Missing values            : 0')
print(f'Columns                   : {df.shape[1]}')
print('='*40)
print('Notebook 01 Complete ')
```

#  IQR Outlier Detection

```python
print('--- Outlier Detection using IQR Method ---\n')

numerical_cols = ['carat', 'price', 'x', 'y', 'z']
outlier_summary = []

for col in numerical_cols:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_summary.append({
        'Column': col,
        'Q1': Q1, 'Q3': Q3, 'IQR': IQR,
        'Lower Bound': lower,
        'Upper Bound': upper,
        'Outlier Count': len(outliers)
    })
    print(f'{col:<8} | IQR: {IQR:.3f} | Lower: {lower:.3f} | Upper: {upper:.3f} | Outliers: {len(outliers)}')

outlier_df = pd.DataFrame(outlier_summary)
```

# Boxplots to Visualize Outliers

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 5))

for i, col in enumerate(['carat', 'price', 'x', 'y', 'z']):
    axes[i].boxplot(df[col], patch_artist=True,
                    boxprops=dict(facecolor='steelblue', color='black'),
                    medianprops=dict(color='red', linewidth=2))
    axes[i].set_title(f'{col}')
    axes[i].set_xlabel(col)

plt.suptitle('Boxplots — Outlier Visualization', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Cap Outliers using IQR

```python
# Cap outliers instead of dropping — preserves data size
for col in ['carat', 'price', 'x', 'y', 'z']:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    before = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    df[col] = df[col].clip(lower=lower, upper=upper)
    print(f'{col:<8} → {before} outliers capped ')

print('\nShape after capping:', df.shape)
```

# Resave Clean Dataset

```python
df.to_csv('../data/diamonds_clean.csv', index=False)
print('Updated clean dataset saved ')
```
