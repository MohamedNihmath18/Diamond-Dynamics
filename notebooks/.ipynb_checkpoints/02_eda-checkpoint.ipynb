# Import Libraries & Load Clean Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/diamonds_clean.csv')
print('Data loaded:', df.shape)
df.head()
```

# Distribution of Price

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Raw price distribution
axes[0].hist(df['price'], bins=50, color='steelblue', edgecolor='black')
axes[0].set_title('Price Distribution (Raw)')
axes[0].set_xlabel('Price (USD)')
axes[0].set_ylabel('Frequency')

# Log transformed price
axes[1].hist(np.log1p(df['price']), bins=50, color='coral', edgecolor='black')
axes[1].set_title('Price Distribution (Log Transformed)')
axes[1].set_xlabel('Log(Price)')
axes[1].set_ylabel('Frequency')

plt.suptitle('Diamond Price Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

#  Distribution of Carat

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['carat'], bins=50, color='mediumseagreen', edgecolor='black')
axes[0].set_title('Carat Distribution (Raw)')
axes[0].set_xlabel('Carat')
axes[0].set_ylabel('Frequency')

axes[1].hist(np.log1p(df['carat']), bins=50, color='orchid', edgecolor='black')
axes[1].set_title('Carat Distribution (Log Transformed)')
axes[1].set_xlabel('Log(Carat)')
axes[1].set_ylabel('Frequency')

plt.suptitle('Diamond Carat Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Count Plots for Categorical Features

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Cut
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.countplot(data=df, x='cut', order=cut_order, palette='Blues_d', ax=axes[0])
axes[0].set_title('Cut Distribution')
axes[0].set_xlabel('Cut')

# Color
color_order = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
sns.countplot(data=df, x='color', order=color_order, palette='Greens_d', ax=axes[1])
axes[1].set_title('Color Distribution')
axes[1].set_xlabel('Color')

# Clarity
clarity_order = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']
sns.countplot(data=df, x='clarity', order=clarity_order, palette='Oranges_d', ax=axes[2])
axes[2].set_title('Clarity Distribution')
axes[2].set_xlabel('Clarity')

plt.suptitle('Categorical Feature Distributions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Price vs Cut, Color, Clarity (Boxplots)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Price vs Cut
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.boxplot(data=df, x='cut', y='price', order=cut_order, palette='Blues', ax=axes[0])
axes[0].set_title('Price vs Cut')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Price (USD)')

# Price vs Color
color_order = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
sns.boxplot(data=df, x='color', y='price', order=color_order, palette='Greens', ax=axes[1])
axes[1].set_title('Price vs Color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Price (USD)')

# Price vs Clarity
clarity_order = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']
sns.boxplot(data=df, x='clarity', y='price', order=clarity_order, palette='Oranges', ax=axes[2])
axes[2].set_title('Price vs Clarity')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Price (USD)')

plt.suptitle('Price Variation by Categorical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(10, 7))

numerical_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
corr_matrix = df[numerical_cols].corr()

sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5)

plt.title('Correlation Heatmap — Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

#  Carat vs Price Scatterplot

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='carat', y='price', 
                hue='cut', 
                palette='viridis', 
                alpha=0.4, 
                s=20)

plt.title('Carat vs Price by Cut', fontsize=14, fontweight='bold')
plt.xlabel('Carat')
plt.ylabel('Price (USD)')
plt.legend(title='Cut', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Average Price per Category (Bar Plots)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Avg price by cut
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
avg_cut = df.groupby('cut')['price'].mean().reindex(cut_order)
axes[0].bar(avg_cut.index, avg_cut.values, color='steelblue', edgecolor='black')
axes[0].set_title('Avg Price by Cut')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Avg Price (USD)')

# Avg price by color
color_order = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
avg_color = df.groupby('color')['price'].mean().reindex(color_order)
axes[1].bar(avg_color.index, avg_color.values, color='mediumseagreen', edgecolor='black')
axes[1].set_title('Avg Price by Color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Avg Price (USD)')

# Avg price by clarity
clarity_order = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']
avg_clarity = df.groupby('clarity')['price'].mean().reindex(clarity_order)
axes[2].bar(avg_clarity.index, avg_clarity.values, color='coral', edgecolor='black')
axes[2].set_title('Avg Price by Clarity')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Avg Price (USD)')

plt.suptitle('Average Price by Category', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Pairplot

```python
# Sample 2000 rows to keep it fast
df_sample = df[['carat', 'x', 'y', 'z', 'price']].sample(2000, random_state=42)

sns.pairplot(df_sample, diag_kind='kde', plot_kws={'alpha': 0.3, 's': 15})
plt.suptitle('Pairwise Relationships — carat, x, y, z, price', 
             fontsize=13, fontweight='bold', y=1.02)
plt.show()
```

# Skewness Check

```python
numerical_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

print('--- Skewness of Numerical Features ---\n')
for col in numerical_cols:
    skew = df[col].skew()
    if abs(skew) > 1:
        flag = ' Highly Skewed'
    elif abs(skew) > 0.5:
        flag = ' Moderately Skewed'
    else:
        flag = ' Normal'
    print(f'{col:<10} skewness: {skew:>8.4f}  {flag}')
```

# Carat vs Price Regression Lineplot

```python
import numpy as np

plt.figure(figsize=(10, 6))

# Sample 3000 points to keep it clean
df_sample = df.sample(3000, random_state=42)

# Scatterplot
plt.scatter(df_sample['carat'], df_sample['price'], 
            alpha=0.3, s=15, color='steelblue', label='Diamonds')

# Regression line
z = np.polyfit(df_sample['carat'], df_sample['price'], 1)
p = np.poly1d(z)
x_line = np.linspace(df_sample['carat'].min(), df_sample['carat'].max(), 100)
plt.plot(x_line, p(x_line), color='red', linewidth=2, label='Regression Line')

plt.title('Carat vs Price — Regression Lineplot', fontsize=14, fontweight='bold')
plt.xlabel('Carat')
plt.ylabel('Price (USD)')
plt.legend()
plt.tight_layout()
plt.show()

# Correlation value
corr_val = df['carat'].corr(df['price'])
print(f'Correlation between Carat and Price: {corr_val:.4f}')
```
