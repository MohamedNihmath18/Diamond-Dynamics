# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('../data/diamonds_engineered.csv')
print('Data loaded:', df.shape)
print('MLPRegressor (ANN) ready ')
```

# Prepare Data

```python
features = ['log_carat', 'cut_encoded', 'color_encoded', 'clarity_encoded',
            'depth', 'table', 'x', 'y', 'z', 'volume', 
            'price_per_carat', 'dimension_ratio']

target = 'log_price_inr'

X = df[features]
y = df[target]

scaler = joblib.load('../models/scaler.pkl')
X_scaled = scaler.transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

print('Train:', X_train.shape)
print('Test :', X_test.shape)
```

#  Build & Train ANN (MLPRegressor)

```python
# MLP = Multi Layer Perceptron = Artificial Neural Network
ann = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),  # 3 hidden layers
    activation='relu',
    solver='adam',
    max_iter=100,
    early_stopping=True,
    validation_fraction=0.1,
    random_state=42,
    verbose=True
)

print('Training ANN model...')
ann.fit(X_train, y_train)
print('Training complete ')
```

# Evaluate ANN

```python
y_pred_ann = ann.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred_ann)
mse  = mean_squared_error(y_test, y_pred_ann)
rmse = np.sqrt(mse)
r2   = r2_score(y_test, y_pred_ann)

print('--- ANN (MLPRegressor) Results ---')
print(f'MAE  : {mae:.4f}')
print(f'MSE  : {mse:.4f}')
print(f'RMSE : {rmse:.4f}')
print(f'R²   : {r2:.4f}')
```

#  Plot Training Loss Curve

```python
plt.figure(figsize=(10, 5))
plt.plot(ann.loss_curve_, label='Training Loss', color='steelblue')
plt.plot(ann.validation_scores_, label='Validation Score', color='coral')
plt.title('ANN Training Loss Curve', fontsize=14, fontweight='bold')
plt.xlabel('Iterations')
plt.ylabel('Loss / Score')
plt.legend()
plt.tight_layout()
plt.show()

print(f'Total iterations ran: {ann.n_iter_}')
```

#  Save ANN Model

```python
joblib.dump(ann, '../models/ann_model.pkl')
print('ANN model saved ')
print('Location: ../models/ann_model.pkl')
```

```python
all_results = pd.DataFrame([
    {'Model': 'Random Forest', 'MAE': 0.003618, 'RMSE': 0.013286, 'R2': 0.999824},
    {'Model': 'XGBoost',       'MAE': 0.009564, 'RMSE': 0.015717, 'R2': 0.999754},
    {'Model': 'Decision Tree', 'MAE': 0.005785, 'RMSE': 0.018783, 'R2': 0.999648},
    {'Model': 'KNN',           'MAE': 0.065602, 'RMSE': 0.092921, 'R2': 0.991390},
    {'Model': 'Linear Reg',    'MAE': 0.070226, 'RMSE': 0.110858, 'R2': 0.987745},
    {'Model': 'ANN (MLP)',     'MAE': 0.043000, 'RMSE': 0.355700, 'R2': 0.873800},
])

print('='*55)
print('        FINAL MODEL COMPARISON')
print('='*55)
print(all_results.sort_values('R2', ascending=False).to_string(index=False))
print('='*55)
print('Best Model: Random Forest ')
```
